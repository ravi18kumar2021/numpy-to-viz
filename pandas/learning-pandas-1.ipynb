### Handling missing data

```python
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eva'],
    'Department': ['HR', 'IT', np.nan, 'Finance', 'IT'],
    'Salary': [50000, np.nan, 60000, 55000, np.nan],
    'Experience': [5, 3, 4, np.nan, 2]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
# Detect missing data
df.isnull().sum()
```

```python
# drop rows containing any missing data
df.dropna()
```

```python
# drop only if salary is missing
df.dropna(subset=['Salary'])
```

```python
# fill missing values with 0
df.fillna(0)
```

```python
# fill missing values with mean/median/mode
df['Department'] = df['Department'].fillna(df['Department'].mode()[0])
df['Salary'] = df['Salary'].fillna(df['Salary'].median())
df['Experience'] = df['Experience'].fillna(df['Experience'].mean())
```

```python
df
```

### Grouping and Aggregation

```python
data = {
    'Flight': ['AA101', 'AA102', 'BA201', 'BA202', 'AA103', 'BA203', 'DL301', 'DL302'],
    'Airline': ['American', 'American', 'British', 'British', 'American', 'British', 'Delta', 'Delta'],
    'Destination': ['NYC', 'LAX', 'NYC', 'LAX', 'LAX', 'NYC', 'LAX', 'NYC'],
    'Delay_Minutes': [30, 0, 45, 10, 60, 5, 20, 15]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
# average delay per airline
df.groupby('Airline')['Delay_Minutes'].mean()
```

```python
# Total delay for each destination
df.groupby('Destination')['Delay_Minutes'].sum()
```

```python
# Counts of flights per airline
df.groupby('Airline')['Flight'].count()
```

```python
df['Airline'].value_counts()
```

```python
# Average delay per airline per destination
df.groupby(['Airline', 'Destination'])['Delay_Minutes'].mean()
```

```python
# Aggregate delay info per airline
df.groupby('Airline')['Delay_Minutes'].agg(['count', 'mean', 'min', 'max', 'sum'])
```

### Merging and Joining DataFrames

```python
theaters = pd.DataFrame({
    'Theater_ID': [1, 2, 3, 4],
    'Theater_Name': ['CineMax', 'MovieWorld', 'Filmplex', 'StarCinema'],
    'City': ['NYC', 'LA', 'NYC', 'Chicago']
})
theaters
```

```python
sales = pd.DataFrame({
    'Sale_ID': [101, 102, 103, 104, 105, 106],
    'Theater_ID': [1, 2, 1, 3, 2, 5],
    'Tickets_Sold': [150, 200, 180, 100, 90, 120],
    'Movie': ['Fast X', 'Oppenheimer', 'Spiderman', 'Barbie', 'Avatar', 'Inception']
})
sales
```

```python
# inner join (only matching Theater_IDs)
pd.merge(left=theaters, right=sales, on='Theater_ID', how='inner')
```

```python
# left join (all theaters, match sales if any)
pd.merge(left=theaters, right=sales, on='Theater_ID', how='left')
```

```python
# right join (all sales, match theaters if any)
pd.merge(left=theaters, right=sales, on='Theater_ID', how='right')
```

```python
# outer join (all data from both sides)
pd.merge(left=theaters, right=sales, on='Theater_ID', how='outer')
```
